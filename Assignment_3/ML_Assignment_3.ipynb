```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_path = "C:\\Users\\saksh\\Desktop\\ml_jupyter_lab\\dataset\\HeartDisease.csv"
df = pd.read_csv(dataset_path)
print("Head of the df: ")
print(df.head())
```

```python
# Bar Plot: ChestPain Frequency
plt.figure(figsize=(8, 6))
df['ChestPain'].value_counts().plot(kind='bar', color='pink')
plt.title('Bar Plot: ChestPain Frequency')
plt.xlabel('ChestPain Type')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot: Age by AHD
plt.figure(figsize=(8, 6))
sns.boxplot(x='AHD', y='Age', data=df)
plt.title('Box Plot: Age Distribution by AHD')
plt.xlabel('AHD')
plt.ylabel('Age')
plt.show()
```

```python
# Scatter Plot: Age vs. MaxHR
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='MaxHR', hue='AHD', data=df)
plt.title('Scatter Plot: Age vs. MaxHR')
plt.xlabel('Age')
plt.ylabel('MaxHR')
plt.show()
```

```python
# Line Chart: Age vs. RestBP
arr_Age = np.array(df["Age"])
sorted_Age = np.sort(arr_Age)
plt.figure(figsize=(8, 6))
plt.plot(sorted_Age, df['RestBP'], marker='o', linestyle='-', color='b')
plt.title('Line Chart: Age vs. RestBP')
plt.xlabel('Age')
plt.ylabel('RestBP')
plt.grid(True)
plt.show()
```

```python
# Pie Chart: Proportion of AHD
plt.figure(figsize=(8, 6))
df['AHD'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Pie Chart: Proportion of AHD')
plt.ylabel('')
plt.show()
```
